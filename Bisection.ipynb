```python
import numpy as np
import math 
import matplotlib.pyplot as plt

def bisection(f,a,b,prc,Nmax):
  """
  Function to find the roots of f(x)..i.e f(x)=0
  using bisection method 
  INPUTS:
  1. Function , f
  2. Left End value , a  (a<b)
  3. Right eng value, b   (f(a)*f(b)<0)
  4. precision , prc
  5. maximum number of iterations, Nmax

  we need to find a value of c such that f(c)=0 
  or which satisfies |f(c)|< prc

  """
  if b<=a:
    print("a must be greater than b")
  elif f(a)*f(b)>=0:
    print("roots cannot be found in this interval")
  else:
    n=1;
    Arr=[];
    fun_c=[];
    while n<= Nmax:
      c=(a+b)*0.5;
      Arr.append(n);
      fun_c.append(f(c));
      if f(c)== 0 or abs (f(c))<prc:
        break
      if f(c)*f(a)>0:
        a=c
      elif f(c)*f(b)>0:
        b=c
      n+=1;
    if n>Nmax:
        print("Maximum number of iteration executed")
    else:
      return Arr, fun_c 

f1 = lambda x: x**3 - 3*(x**2) - x + 9
a1 = -3
b1 = 1

f2 = lambda x: np.exp(x)*(x**3 - 3*(x**2) - x + 9)
a2 = -4
b2 = 4
prc = 1e-10
Nmax = 2000

X1,Y1 = bisection(f1,a1,b1,prc,Nmax)
fig1 = plt.figure()
plt.plot(X1,Y1)
plt.xlabel('number of iterations N')
plt.ylabel('f1(x)')
plt.title('bisection method for function 1')
plt.grid()
plt.tight_layout()
plt.xticks()
plt.yticks()
plt.show()


X2,Y2 = bisection(f2,a1,b1,prc,Nmax)
fig2 = plt.figure()
plt.plot(X2,Y2)
plt.xlabel('number of iterations N')
plt.ylabel('f2(x)')
plt.title('bisection method for function 2')
plt.grid()
plt.tight_layout()
plt.xticks()
plt.yticks()
plt.show()
```
